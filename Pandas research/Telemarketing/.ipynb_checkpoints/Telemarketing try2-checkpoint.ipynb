```python
df1 = pd.read_csv('C:/Users/79295/Desktop/GITHUB/Analytical-practice/datasets/tm_sales_1.csv', sep=';')
df2 = pd.read_csv('C:/Users/79295/Desktop/GITHUB/Analytical-practice/datasets/tm_sales_2.csv', sep=';')
df3 = pd.read_csv('C:/Users/79295/Desktop/GITHUB/Analytical-practice/datasets/tm_sales_3.csv', sep=';')
activations = pd.read_csv('C:/Users/79295/Desktop/GITHUB/Analytical-practice/datasets/prod_activations_logs.csv', sep=';')
```

Суть задачи в том, чтобы проверить подключения продуктов определенным пользователям, соединив файлы о продажах с логами по подключениям в системе.

Особенности данных:

сотрудники телемаркетинга не всегда указывают полный id, если 'id' нет в начале SUBS_ID, то нужно его добавить
поля в файлах могут быть расположены абсолютно случайным образом, но названия полей статичны
продажа не засчитывается, если отключение (END_DTTM) произошло меньше чем через 5 минут после подключения (START_DTTM)
если в файле с продажами встречается строка без указанного SUBS_ID, она пропускается
Сохраните результат в csv файл с разделителем ;, содержащий корректные подключения.

Note: обратите внимание на то, как pandas переводит дату из строки, возможно вам понадобится параметр format

Пример содержания итогового файла (колонка difference это разница между START_DTTM и END_DTTM):

ACT_DTTM;FILIAL_ID;difference 20-03-2020 14:59;1;255 days 09:01:00

```python
df_all = pd.concat([df1, df2])
```

```python
df_all
```

```python
df_all = pd.concat([df_all,df3])
df_all
```

```python
df_all = df_all.dropna()
```

```python
df_all.dtypes
```

```python
df_all.SUBS_ID = df_all.SUBS_ID.apply(lambda x: "id" + x if x.startswith("id") ==False else x)
```

```python
df_all
```

```python
activations
```

```python
activations.START_DTTM = pd.to_datetime(activations.START_DTTM, format = '%d-%m-%Y %H:%M')
activations.END_DTTM = pd.to_datetime(activations.END_DTTM, format = '%d-%m-%Y %H:%M')
```

```python
activations
```

```python
activations['active_time'] = activations.END_DTTM - activations.START_DTTM
```

```python
activations
```

```python
correct_activations = activations[activations.active_time > pd.to_timedelta(5,'m')]
```

```python
correct_activations
```

```python
total = df_all.merge(correct_activations, on = ['SUBS_ID','PROD_ID'])
total
```

```python
total.to_csv('total_correct_activations')
```
