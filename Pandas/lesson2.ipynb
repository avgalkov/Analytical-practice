### Определим какое кол-во взрослых посетителей было по месяцам :

```python
df.groupby('arrival_date_month').aggregate({'adults': 'sum'})
```

Определим какое кол-во детей посетителей было по месяцам в отелях и отсортируем данные по убыванию :

```python
df.groupby(['arrival_date_month', 'hotel']) \
    .aggregate({'children': 'sum'}) \
    .sort_values('children', ascending = False)
```

Сохраним табличку с наши результатом в переменную:

```python
childrens_by_hotels = df.groupby(['arrival_date_month', 'hotel']) \
    .aggregate({'children': 'sum'}) \
    .sort_values('children', ascending = False)
```

Теперь выгружаем все в файл csv:

```python
childrens_by_hotels.to_csv('childrens_by_hotels.csv')
```

На всякий случай проверим сколько у нас отолей в исходном датафрейме:

И удостоверимся, что кол-во детей в нашем получившемся датафрейме совпадает с количеством в исходной таблице и мы никого не потеряли

```python
if childrens_by_hotels.children.sum() == df.children.sum():
    print ('OK')
else:
    print('Error!!!')
```
